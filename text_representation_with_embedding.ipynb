### N-grams

```python
df = pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comment",
                          "dswithbappy write comment"],"output":[1,1,0,0]})

df
```

```python

# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
#Ti gram
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
import numpy as np
import pandas as pd
import gensim
import os
```

```python

from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt_tab')
```

```python
## download the got dataset
import kagglehub

# Download latest version
path = kagglehub.dataset_download("khulasasndh/game-of-thrones-books")

print("Path to dataset files:", path)
```

```python
data = path
```

```python
print(data)
```

```python
## file path
files = os.listdir(data)
files
# first_book = os.path.join(path,'004ssb.txt')
```

```python
new_path = os.path.join(path,'004ssb.txt')
```

```python
story = []
f = open(os.path.join(path,'004ssb.txt'), encoding='cp1252')
corpus = f.read()
raw_sent = sent_tokenize(corpus)
for sent in raw_sent:
  story.append(simple_preprocess(sent))
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10,
    min_count=2
)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('daenerys')
```

```python
model.wv.similarity('arya','sansa')
```
